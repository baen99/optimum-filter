```python
fs = 6.25e5 # sampling frequency [Hz]
"""
Rshunt = 5.0e-3 # shunt resistance
Rfb = 5000.0 # feedback resistance
ADCperVolt = 65536.0#/8 # the number of ADC bins per V for the readout
LoopRatio = 2.4 # SQUID turn ratio 

#scale = 1/(Rfb*ADCperVolt*LoopRatio)
scale = 1/(8 * 12000 * 8192)
"""

drivergain = np.unique(file['detectorConfigDir/detectorConfigZip2/PAS1driverGain'].array(library='np'))[0]
ADCperVolt = np.unique(file['infoDir/userSettingsTree/P_DigitizerBinsPerVolt'].array(library='np'))[0]
FB_gain = np.unique(file['infoDir/userSettingsTree/P_FBgain'].array(library='np'))[0]
loop_gain = np.unique(file['infoDir/userSettingsTree/P_LPGain'].array(library='np'))[0]

scale = 1/(drivergain * ADCperVolt * FB_gain *  loop_gain)
```

```python
sample = 0
channel = 'PFS1'

sample_trace = traces[sample]['Z2'][channel]*scale*1e6 # [A]
sample_trace -= np.mean(sample_trace[0:10000]) # subtract baseline offset

sample_template = np.roll(templates[channel], int(OFdelays[channel][cut][:samplesize][sample]*fs)) # shift template along time axis by RQ OFdelay
sample_template *= OFamps[channel][cut][:samplesize][sample]*1e6 # scale template to RQ OFamp

plt.plot(sample_trace, label='Raw Trace') # UNITS!
plt.plot(sample_template, label='OF from RQs')
plt.title(f'$\chi^2={OFchisq[channel][cut][:samplesize][sample]:.2f}$')
plt.legend()
plt.grid()
plt.ylabel('A')
```

```python
# loop over all 10 traces in all channels and compare OF from RQs to OF from QETpy

OF_qet = {}
OF_rq = {}

for chan in channames:
    
    template = templates[chan]
    psd_ = psd[chan]**2 # has to be in A^2/Hz for OptimumFilter class
    
    OF_qet_tracewise = {}
    OF_rq_tracewise = {}

    for tr in range(len(traces)):
        
        trace = traces[tr]['Z2'][chan]*scale # [A]
        trace -= np.mean(trace[0:10000]) # subtract baseline offset
        
        OF = qp.OptimumFilter(trace, template, psd_, fs) # initialize the OptimumFilter class
        OF_qet_tracewise[evt_num[tr]] = OF.ofamp_withdelay() # calculate amplitude, time shift and chisquare
        # for sake of comparison put RQ values into same form of dictionary
        OF_rq_tracewise[evt_num[tr]] = (OFamps[chan][cut][:samplesize][tr], OFchisq[chan][cut][:samplesize][tr])
        
    OF_qet[chan] = OF_qet_tracewise
    OF_rq[chan] = OF_rq_tracewise
    
    
    
```
